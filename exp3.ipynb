```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv("data.csv")
print(a)
plt.hist(a['price'], bins=20, color='orange', edgecolor='black')
plt.xlabel('Price')
plt.title('Histogram of House Prices') 
plt.show()






```

```python

x= a['bathrooms']
y = a['price']
plt.xlabel("bathrooms")
plt.ylabel("price")
plt.bar(x,y,color ='red')
plt.bar(x,y, color ='blue')
plt.show()
```

```python
b = a.head(300)
plt.boxplot(b['sqft_living'])
plt.xlabel('sqft living')
plt.title('Boxplot of sqft living')
plt.show()
```

```python
x = a['bathrooms']
y = a['price']
plt.xlabel("price")
plt.ylabel("bathroom")
plt.scatter(x,y )
```

```python
b = a.head(5)
y = b['bathrooms']
x = b['price']
plt.xlabel("price")
plt.ylabel("bathroom")
plt.plot(x, y, marker='o', linestyle='-', color='b')

# 
```

```python
import seaborn as sns
b = a.head(10)
numeric_data = b.select_dtypes(include =('number'))
df = numeric_data.corr()
sns.heatmap(df, annot=True,cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of DataFrame')
plt.show()  
```
